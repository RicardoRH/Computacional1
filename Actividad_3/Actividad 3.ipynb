```python
#Primer indicación: cargar a la memoria las bibliotecas Pandas (manejo de datos, Numpy (numerical python) 
#y la biblioteca de gráficas Matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Lee tu archivo de datos a tu sesión de trabajo y asígnalo a un data frame.
#Datos correspondientes al 22 de junio.
df1 = pd.read_csv('BrindisiJune.txt', skiprows=5, sep='\s+')
#Ahora, nombraremos las columnas.
df1.columns = ['PRES','HGHT','TEMP','DWPT','RELH','MIXR','DRCT','SKNT','THTA','THTE','THTV']
#Datos correspondientes al 22 de diciembre.
df2 = pd.read_csv('BrindisiDec.txt', skiprows=5, sep='\s+')
#Ahora, nombraremos las columnas.
df2.columns = ['PRES','HGHT','TEMP','DWPT','RELH','MIXR','DRCT','SKNT','THTA','THTE','THTV']
```

```python
#Transformamos los objetos a numero
columnas = ['PRES','HGHT','TEMP','DWPT','RELH','MIXR','DRCT','SKNT','THTA','THTE','THTV']
df1[columnas]= df1[columnas].apply(pd.to_numeric,errors='coerce', axis=1)
df2[columnas]= df2[columnas].apply(pd.to_numeric,errors='coerce', axis=1)

```

```python
df1.dtypes
```

```python
#Gráfica de la variación de la presión con respecto a la altura del mes de junio.
df1P = df1[['HGHT','PRES']]
plt.figure(); df1P.plot(x='PRES');plt.legend(loc='best')
plt.title('Variación de la Presión con Respecto a la Altura de junio')
plt.xlim([0,1010])
plt.xlabel('Presión (hPa)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()
```

```python
#Gráfica la variación de la presión con respecto a la altura del mes diciembre.
df2P = df2[['HGHT','PRES']]
plt.figure(); df2P.plot(x='PRES');plt.legend(loc='best')
plt.title('Variación de la Presión con Respecto a la Altura de diciembre')
plt.xlim([0,1025])
plt.xlabel('Presión (hPa)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()
```

```python
#Gráfica la variación de la temperatura con respecto a la altura del mes de junio.
df1T = df1[['HGHT','TEMP']]
plt.figure(); df1T.plot(x='TEMP');plt.legend(loc='best')
plt.title('Variación de la Temperatura con Respecto a la Altura de junio')
plt.xlim([-70,10])
plt.xlabel('Temperatura (°C)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()
```

```python
#Gráfica la variación de la temperatura con respecto a la altura del mes de diciembre.
df2T = df2[['HGHT','TEMP']]
plt.figure(); df2T.plot(x='TEMP');plt.legend(loc='best')
plt.title('Variación de la Temperatura con Respecto a la Altura de junio')
plt.xlim([-70,10])
plt.xlabel('Temperatura (°C)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()
```

```python
#Gráfica de la temperatura y temperatura de rocío, como función de la altura del mes de junio.
x1=df1['TEMP']
x2=df1['DWPT']
y=df1['HGHT']
plt.plot(x1, y, label='TEMP')
plt.plot(x2, y, label='DWPT')
plt.legend(loc='best')
plt.title("Gráfica de la Temperatura y Temperatura de Rocío Respecto a la Altura de junio")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Altura (m)")
plt.grid(True)
plt.show()
```

```python
#Gráfica de la temperatura y temperatura de rocío, como función de la altura del mes de diciembre.
x1=df2['TEMP']
x2=df2['DWPT']
y=df2['HGHT']
plt.plot(x1, y, label='TEMP')
plt.plot(x2, y, label='DWPT')
plt.legend(loc='best')
plt.title("Gráfica de la Temperatura y Temperatura de Rocío Respecto a la Altura de junio")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Altura (m)")
plt.grid(True)
plt.show()
```

```python
#Gráfica de la rapidez de los vientos en nudos de junio.
df1N = df1[['SKNT','HGHT']]
plt.figure(); df1N.plot(x='SKNT'); plt.legend(loc='best')
plt.title("Rapidez de los Vientos en Nudos de junio")
plt.xlabel("Rapidez (Nudos)")
plt.ylabel('Altura (m)')
plt.xlim([0,200])
plt.grid(True)
plt.show()
```

```python
#Gráfica de la rapidez de los vientos en nudos de diciembre.
df2N = df2[['SKNT','HGHT']]
plt.figure(); df2N.plot(x='SKNT'); plt.legend(loc='best')
plt.title("Rapidez de los Vientos en Nudos de diciembre")
plt.xlabel("Rapidez (Nudos)")
plt.ylabel('Altura (m)')
plt.xlim([0,200])
plt.grid(True)
plt.show()
```

```python
#Gráfica de la humedad relativa como función de la altura de junio.
df1H = df1[['RELH','HGHT']]
plt.figure(); df1H.plot(x='RELH');plt.legend(loc='best')
plt.title('Humedad Relativa Respecto de la Altura de junio')
plt.xlabel('Humedad Relativa (%)')
plt.ylabel('Altura (m)')
plt.xlim([0,100])
plt.grid(True)
plt.show()
```
